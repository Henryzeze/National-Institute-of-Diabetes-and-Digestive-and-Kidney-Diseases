###Identify and handle missing values

```python
missing_data = df.isnull()
missing_data.head() # "True" stands for missing value, while "False" stands for not missing value.
```

###Count missing values in each column

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

###Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labels= 'Not Diabetic','Diabetic'
plt.pie(df['Outcome'].value_counts(),labels=labels,autopct='%0.02f%%')
plt.legend()
plt.show()
```

As you can see above, 65.10% females are not Diabetic and 34.90% are Diabetic.

```python
# To visualize correlation using a Heatmap
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
